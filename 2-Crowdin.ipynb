```python
# %pip install crowdin-api-client
import os
from dotenv import load_dotenv
from crowdin_api import CrowdinClient

load_dotenv()

CROWDIN_PROJECT_ID=393919
client = CrowdinClient(token=os.getenv('CROWDIN_TOKEN'), project_id=CROWDIN_PROJECT_ID, )
```

```python
import pandas as pd
df=pd.read_pickle('cleaned_quests.pkl')
df.head()
```

```python
# %pip install ratelimit
from datetime import timedelta
from ratelimit import limits, sleep_and_retry
# from threading import Semaphore

# semaphore = Semaphore(20)

@sleep_and_retry
@limits(calls=19, period=timedelta(seconds=1).total_seconds())
def query_glossary_api(row):
    # with semaphore:
    texts = [row['title_en'], row['objective_en'], row['description_en'], row['progress_en'], row['completion_en']]
    texts = [text for text in texts if text is not None]
    # print(texts)
    glossary_response = client.glossaries.concordance_search_in_glossaries(sourceLanguageId='en', targetLanguageId='uk', expressions=texts)
    # terms = list(map(lambda data: (data['data']['sourceTerms'][0]['text'], data['data']['targetTerms'][0]['text'], data['data']['sourceTerms'][0]['description']), glossary_response['data']))
    terms = list(map(lambda data: {'en': data['data']['sourceTerms'][0]['text'], 'uk': data['data']['targetTerms'][0]['text']}, glossary_response['data']))
    return terms
    
```

```python
from concurrent.futures import ThreadPoolExecutor
with ThreadPoolExecutor(max_workers=20) as executor:
    df['glossary'] = list(executor.map(query_glossary_api, [row for _, row in df.iterrows()]))
```
