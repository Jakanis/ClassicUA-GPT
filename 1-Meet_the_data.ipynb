```python
import sqlite3
import pandas as pd

with sqlite3.connect('input/quests.db') as conn:
    df = pd.read_sql_query("SELECT * FROM quests", conn)

df.head()
```

```python
df.info()
```

```python
def generate_json(row, fields):
    return row[fields].to_json(force_ascii=False)

def generate_en_json(row):
    return generate_json(row, ["title_en", "objective_en",	"description_en",	"progress_en",	"completion_en"])

def generate_uk_json(row):
    return generate_json(row, ["title_uk", "objective_uk",	"description_uk",	"progress_uk",	"completion_uk"])

generate_en_json(df.iloc[0])
```

```python
generate_uk_json(df.iloc[0])
```

```python
non_translated = df.loc[df['title_uk'].isnull()]
non_translated.info()
```

```python
def save_dataframe(df, filename=None):
    df_len = len(df)
    if df_len:
        filename = f'{"DataFrame" if filename is None else filename}_{str(df_len)}.pkl'
        print(df_len, ' len dataset will be saved to ', filename)
        df.to_pickle(filename)
        print('Dataset saved to ', filename)
```

```python
save_dataframe(df, "all_quests")
```
